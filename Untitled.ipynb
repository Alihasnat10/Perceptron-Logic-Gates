# Perceptron Logic Gates
## PERCEPTRONS AND NEURAL NETS
#### In this project, we will use perceptrons to model the fundamental building blocks of computers â€” logic gates.

```python
from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
import numpy as np
from itertools import product
```

```python
data =[[0,0],[0,1],[1,0],[1,1]] #inputs to gates
labels = [0,1,1,1] #outputs
plt.scatter([point[0] for point in data],[point[1] for point in data], c= labels ) #scatter plot
plt.show()
```

```python
classifier = Perceptron(max_iter = 40) #building perceptron
classifier.fit(data, labels)
print(classifier.score(data, labels))
```

```python
#visualizing the perceptron
x_values=np.linspace(0,1,100)
y_values=np.linspace(0,1,100)
point_grid = list(product(x_values,y_values))
distances=classifier.decision_function(point_grid)
abs_distances = abs(distances)
distances_matrix = np.reshape(abs_distances, (100, 100))
```

```python
#Matplotlibâ€™s pcolormesh() function
heatmap = plt.pcolormesh(x_values, y_values, distances_matrix)
plt.colorbar(heatmap)
plt.show()
```
